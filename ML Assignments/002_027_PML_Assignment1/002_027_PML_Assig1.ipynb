```python
### Sagar Raut PRN : 240340128027
### Ameya Panshikar PRN : 240340128002
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import StackingRegressor
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
# Load the training and testing datasets
abalone_train = pd.read_csv(r"D:\PML\Kaggle\RegressionWithAnAbaloneDataset\train.csv\train.csv")
abalone_test = pd.read_csv(r"D:\PML\Kaggle\RegressionWithAnAbaloneDataset\test.csv\test.csv")
```

```python
# Convert categorical variable 'Sex' to numerical values using LabelEncoder
le = LabelEncoder()
abalone_train['Sex'] = le.fit_transform(abalone_train['Sex'])
abalone_test['Sex'] = le.transform(abalone_test['Sex'])
```

```python
# Split the data into features (X) and target (y)
X = abalone_train.iloc[:, 1:-1]
y = abalone_train.iloc[:, -1]
```

```python
# Define the regression models
lr = LinearRegression()
rf = RandomForestRegressor()
x_gbm = XGBRegressor()
l_gbm = LGBMRegressor()
stack = StackingRegressor([('LR', lr), ('XGBM', x_gbm), ('LGBM', l_gbm)], final_estimator=rf, passthrough=True)
```

```python
# Train the models
lr.fit(X, y)
x_gbm.fit(X, y)
l_gbm.fit(X, y)
stack.fit(X, y)
```

```python
# Make predictions on the testing set
y_pred_lr = lr.predict(abalone_test.iloc[:, 1:])
y_pred_xgb = x_gbm.predict(abalone_test.iloc[:, 1:])
y_pred_lgb = l_gbm.predict(abalone_test.iloc[:, 1:])
y_pred_stack = stack.predict(abalone_test.iloc[:, 1:])
```

```python
# Create submission files for each model
submit_lr = pd.DataFrame({'id': abalone_test['id'], 'Rings': y_pred_lr})
submit_lr.to_csv('sbt_abalone_lr.csv', index=False)

submit_xgb = pd.DataFrame({'id': abalone_test['id'], 'Rings': y_pred_xgb})
submit_xgb.to_csv('sbt_abalone_xgbm.csv', index=False)

submit_lgb = pd.DataFrame({'id': abalone_test['id'], 'Rings': y_pred_lgb})
submit_lgb.to_csv('sbt_abalone_lgbm.csv', index=False)

submit_stack = pd.DataFrame({'id': abalone_test['id'], 'Rings': y_pred_stack})
submit_stack.to_csv('sbt_abalone_stack.csv', index=False)
```
