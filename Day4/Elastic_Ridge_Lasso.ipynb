```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split 
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.metrics import accuracy_score, r2_score
from sklearn.naive_bayes import GaussianNB
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import StratifiedKFold, cross_val_score
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.linear_model import LinearRegression, Ridge,ElasticNet, Lasso
from sklearn.model_selection import KFold
```

```python
insurance = pd.read_csv(r'C:\Users\Administrator.DAI-PC2\Desktop\jupiter_demo\ML\Cases-20240426T111123Z-001\Cases\Medical Cost Personal\insurance.csv')
print(insurance.info())
```

```python
dum = pd.get_dummies(insurance, drop_first=True)
dum
```

```python
X = dum.drop('charges',axis=1)  #take all the columns upto the salary column
y = dum['charges']  #take the first column
print(y)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=24)
```

```python
elastic=ElasticNet()
ridge = Ridge()
lasso = Lasso()
lr = LinearRegression()

# print(elastic.get_params())

#linear regression
kfold = KFold(n_splits=5, shuffle=True, random_state=24)
results = cross_val_score(lr,X,y,cv=kfold)
print(results.mean())

#### Elastic Net
params={'alpha':np.linspace(0,100,50),'l1_ratio': np.linspace(0,1,10)}
gcv=GridSearchCV(elastic,param_grid=params,cv=kfold,scoring='r2')
gcv.fit(X,y)
pd_cv= pd.DataFrame(gcv.cv_results_)
print(gcv.best_params_)
print(gcv.best_score_)
```

```python
#### Ridge
params={'alpha':np.linspace(0,100,50)}
gcv=GridSearchCV(ridge,param_grid=params,cv=kfold,scoring='r2')
gcv.fit(X,y)
pd_cv= pd.DataFrame(gcv.cv_results_)
print(gcv.best_params_)
print(gcv.best_score_)
```

```python
#### Lasso
params={'alpha':np.linspace(0,100,50)}
gcv=GridSearchCV(lasso,param_grid=params,cv=kfold,scoring='r2')
gcv.fit(X,y)
pd_cv= pd.DataFrame(gcv.cv_results_)
print(gcv.best_params_)
print(gcv.best_score_)
```
