```python
############## Ridge ###############
import warnings
warnings.filterwarnings("ignore")

ridge=Ridge()
poly=PolynomialFeatures()
pipe= Pipeline([('POLY',poly),('LR',ridge)])
kfold = KFold(n_splits =5, shuffle=True, random_state=24)
print(pipe.get_params())
params={'POLY__degree': [1,2,3,4,5],'LR__alpha':np.linspace(0.001,5,10)}
gcv=GridSearchCV(pipe,param_grid=params,cv=kfold)
gcv.fit(X,y)
print(gcv.best_score_)
print(gcv.best_params_)

        
        
```

```python

#####################  Elastic  $$$$$$$$$$$$$$$$
elastic=ElasticNet()
poly=PolynomialFeatures()
pipe= Pipeline([('POLY',poly),('LR',elastic)])
print(pipe.get_params())
params={'POLY__degree': [1,2,3],'LR__alpha':np.linspace(0.001,5,10),'LR__l1_ratio':np.linspace(0,1,5)}
gcv=GridSearchCV(pipe,param_grid=params,cv=kfold)
gcv.fit(X,y)
print(gcv.best_score_)
print(gcv.best_params_)
```

```python

################     lasso ###################
lasso=Lasso()
poly=PolynomialFeatures()
pipe= Pipeline([('POLY',poly),('LR',lasso)])
print(pipe.get_params())
params={'POLY__degree': [1,2,3],'LR__alpha':np.linspace(0.001,5,10)}
gcv_lass=GridSearchCV(pipe,param_grid=params,cv=kfold)
gcv_lass.fit(X,y)
print(gcv_lass.best_score_)
print(gcv_lass.best_params_)
```
